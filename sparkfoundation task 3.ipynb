# Explolatory data analysis - Retail

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("SampleSuperstore.csv")
data
```

```python
profit_by_category = data.groupby('Category')['Profit'].sum().sort_values(ascending=False)
```

```python
profit_by_region = data.groupby('Region')['Profit'].sum().sort_values(ascending=False)
```

```python
profit_by_segment = data.groupby('Segment')['Profit'].sum().sort_values(ascending=False)
```

```python
profit_by_subcategory = data.groupby('Sub-Category')['Profit'].sum().sort_values(ascending=False)
```

```python
shipping_costs = data.groupby('Ship Mode')['Profit'].sum().sort_values(ascending=False)
sns.set_style("whitegrid")
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=profit_by_category.index, y=profit_by_category.values, palette="Blues_d")
plt.xlabel('Category')
plt.ylabel('Profit')
plt.title('Profit by Category')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=profit_by_region.index, y=profit_by_region.values, palette="Greens_d")
plt.xlabel('Region')
plt.ylabel('Profit')
plt.title('Profit by Region')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=profit_by_segment.index, y=profit_by_segment.values, palette="Reds_d")
plt.xlabel('Segment')
plt.ylabel('Profit')
plt.title('Profit by Customer Segment')
plt.xticks(rotation=45)
plt.show()
```

```python

plt.figure(figsize=(12, 6))
sns.barplot(x=profit_by_subcategory.index, y=profit_by_subcategory.values, palette="Oranges_d")
plt.xlabel('Sub-Category')
plt.ylabel('Profit')
plt.title('Profit by Sub-Category')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=shipping_costs.index, y=shipping_costs.values, palette="Purples_d")
plt.xlabel('Ship Mode')
plt.ylabel('Profit')
plt.title('Profit by Ship Mode')
plt.xticks(rotation=45)
plt.show()
```
